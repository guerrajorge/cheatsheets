```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
# `iris.data` holds the numerical values
# `iris.feature_names` holds the numerical column names
# `iris.target` holds the categorical (species) values (as ints)
# `iris.target_names` holds the unique categorical names
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris['feature_names'])
```

Info on datafram

```python
df.info()
```

Summary statistics

```python
df.describe()
```

Binning Data

```python
bins = [0, 2, 4, 6]
labels =['short','medium','long']

category = pd.cut(df['petal length (cm)'], bins = bins, labels = labels)
df['category'] = category
df[['petal length (cm)', 'category']].head()
```

Pandas Crosstab

```python
pd.crosstab(df['petal length (cm)'], df["sepal width (cm)"], margins=True, normalize = 'index').head()
```
